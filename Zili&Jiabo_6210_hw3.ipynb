## Step Three: Answering the questions.
#### What are tags are associated with a person, place or thing?
We could query the dataset to get the info we needed. For example: 
Query the hashtags of tweets of 'NintendoAmerica'(A game company name)

```python
from bson.json_util import dumps
pipeline =twi.aggregate(
    [
        {'$match':{'user':'NintendoAmerica'}},
        {'$unwind':'$hashtags'},
        {'$group':{'_id':'$hashtags.text',
                   'total':{'$sum':1}    
        }},
        {'$sort':{'total':-1}}
        
    ]
)
dumps(pipeline)
```

The hashtags most presents of NintendoAmerica is "#AnimalCrossing","# NintendoSwitchOnline"
#### What social media users are like other social media users in your domain?

```python
pipeline =fri.aggregate(
    [
        {'$group':{'_id':'$name',
                   'Freq':{'$sum':1}    
        }},
        {'$sort':{'Freq':-1}}, 
        {'$limit': 14}
    ]
)
dumps(pipeline)
```

The result shows 10 people in the dataset all followed Fortnite, which means the Fortnite is the most likely user in my domain who would be followed.
#### What people, places or things are popular in your domain?
The most popular things in my domain:
Query the hashtags and sort by the total number of their presents:

```python
pipeline=twi.aggregate(
    [        
        {'$unwind':'$hashtags'},
        {'$group':{'_id':'$hashtags.text',
                   'total':{'$sum':1}    
        }},
        {'$sort':{'total':-1}},
        {'$limit': 20}
    ]
)
dumps(pipeline)
```

The popular hashtags includes: COVID19, Fortnite, ThisIsMyStreet, StayHome, AnimalCrossing, SonicWatchParty, ect.
